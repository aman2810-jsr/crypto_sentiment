```python
import numpy as np
import pandas as pd

df_prices=pd.read_csv("../dataset/BTC-Hourly.csv")

df_prices.head()
```

```python
drop_colums=['unix','symbol']
df_prices.drop(columns=drop_colums, inplace=True)

df_prices.head()
```

```python
df_prices['date']=pd.to_datetime(df_prices['date'])

start_time=pd.to_datetime('2021-02-05 10:00:00')
end_time=pd.to_datetime('2021-07-30 23:00:00')

df_prices=df_prices[(df_prices['date'] >= start_time) & (df_prices['date'] <= end_time)]

df_prices = df_prices.reset_index(drop=True)
```

```python
df_prices = df_prices.sort_values(by='date').reset_index(drop=True)
df_prices
```

```python
df_sentiment=pd.read_csv('../dataset/hourly_avg_sentiment.csv')
df_sentiment
```

## Filling the missing timestamps in the df_sentiment with 0 and merging both dataframes

```python
import pandas as pd

# Assuming you already did:
df_prices['date'] = pd.to_datetime(df_prices['date'])
df_sentiment['hour'] = pd.to_datetime(df_sentiment['hour'])

df_prices = df_prices.set_index('date')
df_sentiment = df_sentiment.set_index('hour')

# Create the missingness flag BEFORE filling NaNs
missing_flag = df_sentiment.isna().any(axis=1).astype(int)
# This flag is 1 if any column in that row is NaN, else 0

# Reindex sentiment to match price timestamps
df_sentiment = df_sentiment.reindex(df_prices.index)

# Reindex flag as well, align with sentiment index
missing_flag = missing_flag.reindex(df_prices.index).fillna(1).astype(int)  # missing rows get flag=1

# Fill missing sentiment values with 0
df_sentiment = df_sentiment.fillna({
    'avg_sentiment': 0.0,
    'sentiment_volatility': 0.0,
    'tweet_volume': 0,
    'count_neg': 0,
    'count_neu': 0,
    'count_pos': 0
})

# Add missingness flag column to sentiment dataframe
df_sentiment['sentiment_missing_flag'] = missing_flag

# Merge price and sentiment dataframes
df_merged = pd.concat([df_prices, df_sentiment], axis=1).reset_index()

print(df_merged[['date', 'sentiment_missing_flag']].head(1000))
```

```python
df_merged.head()
```

```python
(df_merged['avg_sentiment']==0).sum()
```

```python
df_merged['returns'] = (df_merged['close'] - df_merged['open']) / df_merged['open']
df_merged['volatility'] = (df_merged['high'] - df_merged['low']) / df_merged['open']
df_merged['momentum_3'] = df_merged['close'] / df_merged['close'].shift(3) - 1
df_merged['rolling_mean_close_3'] = df_merged['close'].rolling(window=3).mean()
df_merged['rolling_std_close_3'] = df_merged['close'].rolling(window=3).std()
df_merged['high_low_ratio'] = df_merged['high'] / df_merged['low']
```

```python
df_merged['future_return'] = (df_merged['close'].shift(-1) - df_merged['close']) / df_merged['close']

def label_action(x, threshold=0.01):
    if x > threshold:
        return 1  # buy
    elif x < -threshold:
        return 2  # sell
    else:
        return 0  # hold

df_merged['target'] = df_merged['future_return'].apply(label_action)

```

```python
df_merged['sentiment_diff'] = df_merged['count_pos'] - df_merged['count_neg']
df_merged['sentiment_ratio'] = (df_merged['count_pos'] + 1) / (df_merged['count_neg'] + 1)
df_merged['sentiment_intensity'] = df_merged['avg_sentiment'] * df_merged['tweet_volume']
```

```python
df_merged['volume_btc_change'] = df_merged['Volume BTC'].pct_change() 
df_merged['volume_usd_change'] = df_merged['Volume USD'].pct_change() 
df_merged['tweet_volume_change'] = df_merged['tweet_volume'].pct_change() 
```

## XGBoost model

```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
import matplotlib.pyplot as plt

features=df_merged.drop(columns=['date','future_return','target'])
target=df_merged['target']

```

```python
inf_columns = features.columns[np.isinf(features).sum() > 0].tolist()
inf_columns
```

```python
np.isinf(features['tweet_volume_change']).sum()
```

```python
mask = ~np.isinf(features['tweet_volume_change'])
features = features[mask].reset_index(drop=True)
target = target[mask].reset_index(drop=True)
```

```python
features.fillna(0, inplace=True)
target.fillna(0, inplace=True)
```

```python
split_index= int(len(df_merged) * 0.8)
X_train = features[:split_index]    
X_test = features[split_index:]
y_train = target[:split_index]
y_test = target[split_index:]

model=XGBClassifier(
    eta=0.1,
    max_depth=6,
    objective='multi:softmax',
    num_class=3,
    eval_metric='mlogloss',
    reg_lambda=2,
    reg_alpha=0.5
)
model.fit(X_train, y_train)
```

```python
predictions= model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)



```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=['Hold', 'Buy', 'Sell']))
```
